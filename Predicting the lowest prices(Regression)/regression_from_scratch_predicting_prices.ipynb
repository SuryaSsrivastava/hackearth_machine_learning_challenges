## 5. preparing our Algorithm

```python
class Model:
    def __init__(self,total):
        self.total=total
        self.n_features = total[1]
        self.w=np.random.randn(total[1],1)
        self.b=np.random.randn()
    
    def forward_pass(self,X):
        return (np.dot(X,self.w)+self.b)
    
    def error(self,y,y_true):
        loss = np.sum(np.square(y - y_true))
        return loss/(2*y.shape[0])
    
    def backward_pass(self, X, y_true, y_hat):
        m = y_hat.shape[0]
        db = np.sum(y_hat - y_true)/m
        dW = np.sum(np.dot((y_hat - y_true).T, X),axis=0)/m
        return dW, db
    
    def gradient_descent(self, dW, db, lr):
        self.w = self.w - lr * np.reshape(dW, (self.n_features, 1))
        self.b = self.b - lr * db
        
    def fit_transform(self,x_train,y_train,max_itr=500,lr=0.001):
        
        losses=[]
        
        for i in range(max_itr):
            y_hat = self.forward_pass(x_train)
            dW, db = self.backward_pass(x_train, y_train, y_hat)
            self.gradient_descent(dW, db, lr)
            loss = self.error(y_hat, y_train)
            losses.append(loss)
            if i % 100 == 0:
                print('Iter: {}, Current loss: {:.4f}'.format(i, loss))
                
        return losses
```
