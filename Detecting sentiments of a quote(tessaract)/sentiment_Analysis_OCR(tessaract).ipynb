### 1. Reading Images from Directory

```python
filepath='./Dataset/'
fig,ax=plt.subplots(5,figsize=(25,25))
for i in range(5):
    img=os.path.join(filepath,os.listdir(filepath)[i])
    img=image.load_img(img)
    img=image.img_to_array(img)/255.
    ax[i].imshow(img)
    ax[i].axis('off')
```
