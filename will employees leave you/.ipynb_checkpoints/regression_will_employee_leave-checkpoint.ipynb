### importing essential libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train=pd.read_csv("./Train.csv")
df_test=pd.read_csv("./Test.csv")
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
df_train.head()
```

```python
df_train.info()
```

### convert catagorical to numerical
- Gender
- Relationship_Status        
- Hometown                   
- Unit                        
- Decision_skill_possess     
- Time_of_service
- Compensation_and_Benefits

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
columns=['Gender',
        'Relationship_Status',
        'Hometown',
        'Unit',
        'Decision_skill_possess',
        'Time_of_service',
        'Compensation_and_Benefits']

for col in columns:
    df_train[col]=le.fit_transform(df_train[col])
```

```python
df_train.head()
```

```python
for col in columns:
    df_test[col]=le.fit_transform(df_test[col])
```

```python
df_test.head()
```

```python
Y_train=df_train['Attrition_rate']
employee_id=df_test['Employee_ID']
```

```python
df_train.drop(columns=['Employee_ID','Attrition_rate'],axis=1,inplace=True)
df_test.drop(columns=['Employee_ID'],axis=1,inplace=True)
```

```python
df_train.head()
```

```python
df_test.head()
```

### filling Nan values

```python
df_train=df_train.fillna(df_train.mean())
df_test=df_test.fillna(df_test.mean())
```

```python
df_train.isnull().sum().sort_values(ascending=False)[:5]
```

```python
df_test.isnull().sum().sort_values(ascending=False)[:5]
```

### normalize the data

```python
df_train=(df_train-df_train.mean())/df_train.std()
df_test=(df_test-df_test.mean())/df_test.std()
#Y_train=(Y_train-Y_train.mean())/Y_train.std()
```

```python
df_train.head(2)
```

```python
X_train=df_train.values
Y_train=Y_train
X_test=df_test.values
```

### using Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
lr=LinearRegression(normalize=True)
```

```python
lr.fit(X_train,Y_train)
```

```python
prediction=lr.predict(X_test)
```

```python
temp=pd.DataFrame({'Employee_ID':employee_id,'Attrition_rate':prediction})
temp.to_csv('linear_regression.csv',index=False)   # 81% accuracy
```

### using decision trees

```python
dtr=DecisionTreeRegressor()
```

```python
dtr.fit(X_train,Y_train)
```

```python
prediction=dtr.predict(X_test)
```

```python
temp=pd.DataFrame({'Employee_ID':employee_id,'Attrition_rate':prediction})
temp.to_csv('decision_trees.csv',index=False)   # 81% accuracy
```

### using gridsearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={
    'ccp_alpha':[0.0,0.1,0.3,0.5,1.0],
    'criterion':["mse", "friedman_mse", "mae"],
#     'max_depth':[10,15,20,30,35,50],
#     'max_leaf_nodes':[10,20,25,30],
#     'min_samples_leaf':[1,2,3,4,5],
#     'min_samples_split':[2,3,4,5],
    'min_weight_fraction_leaf':[0.0,0.5,1.0],
    'splitter':["best", "random"]
}
```

```python
gs=GridSearchCV(dtr,param_grid=params,n_jobs=-1,verbose=1)
```

```python
gs.fit(X_train,Y_train)
```

```python
gs.best_estimator_
```

```python
prediction=gs.predict(X_test)
```

```python
temp=pd.DataFrame({'Employee_ID':employee_id,'Attrition_rate':prediction})
temp.to_csv('grid_search.csv',index=False)   # 81% accuracy
```
